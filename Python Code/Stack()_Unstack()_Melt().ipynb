* Stack allows information from columns to swing to indexes and vice versa (with unstack()). 



* Stack and unstack method works with the MultiIndex objects in DataFrame

* unstack the prescribed level(s) from row to column

```python
import pandas as pd
import numpy as np
```

```python
df= pd.DataFrame([[0, 1], [2, 3]],
                 index=['cat', 'dog'],
                 columns=['weight', 'height'])
```

```python
df
```

```python
stacked=df.stack()
print('*'*30,'stacked dataframe','*'*30)
print(stacked)
print('*'*30,'unstacked dataframe','*'*30)
print(stacked.unstack())
```

```python
stacked.unstack()
```

```python
header = pd.MultiIndex.from_product([['Semester1','Semester2'],['Maths','Science']])

d=([[12,45,67,56],[78,89,45,67],[45,67,89,90],[67,44,56,55]])
df = pd.DataFrame(d,
                  index=['Alisa','Bobby','Cathrine','Jack'],
                  columns=header)
df
```

* When more than one column header is present we can stack the specific column header by specified the level.
* Stacking a dataframe at level 1 will stack maths and science columns row wise
* Stacking a dataframe at level 0 will stack semester1 and semester2 columns row wise.

```python
stacked_level1=df.stack() # or df.stack(level=1) is same
stacked_level1
```

```python
#stacking at different level
stacked_level0=df.stack(level=0)
stacked_level0
```

* unstack() Function in dataframe unstacks the row to columns . Basically itâ€™s a reverse of stacking

```python
unstacked=stacked_level1.unstack()
unstacked
```

```python
unstacked_level0=stacked_level0.unstack()
unstacked_level0
```

* Melt in pandas reshape dataframe from wide format to long format.

* Melt combines columns to a standard 3-column format of id_vars, variable, value to allow columnar processing of variable before pivoting back with edited column values.

* We can use pivot() function to unmelt the output dataframe back to the original dataframe.

```python
d1 = {"Name": ["Nitesh", "Sam", "Tom"], 
      "ID": [11, 22, 33], 
      "Role": ["Developer", "Admin", "CEO"]}

df=pd.DataFrame(d1)
```

```python
df
```

```python
#pandas melt with multiple value_vars
df_melted=pd.melt(df,id_vars=['ID'],value_vars=['Name','Role'])
```

```python
df_melted
```

```python
#pandas melt example
df_melted=pd.melt(df,id_vars=['ID'])
df_melted
```

```python
#Multiple columns as id_vars
pd.melt(df,id_vars=['ID','Name'],value_vars=['Role'])
```
