```python
#Find the Monthly Percentage Difference between purchase.
order_data=pd.DataFrame(purchase_data,columns=cols_name)
```

```python
order_data['created_at']=order_data['created_at'].astype('datetime64[ns]')
```

```python
order_data.info()
```

```python
order_data['created_at'].dt.month.unique
```

```python
order_data['month']=order_data['created_at'].dt.month
```

```python
order_data['date_formatted']=order_data['created_at'].dt.strftime('%Y-%m')
```

```python
order_data
```

```python
month_revenue=order_data.groupby(['date_formatted']).sum().reset_index()[['date_formatted','value']]
```

```python
month_revenue['diff_val']=month_revenue['value'].diff()
month_revenue
```

```python
month_revenue['last_month_revenue']=month_revenue['value']-month_revenue['diff_val']
month_revenue
```

```python
month_revenue['Monthly_pct_diff']=(month_revenue['diff_val']/month_revenue['last_month_revenue'])*100
```

```python
month_revenue['Monthly_pct_diff']=month_revenue['Monthly_pct_diff'].round(2)
month_revenue[['date_formatted','Monthly_pct_diff']]
```
